```python
import pandas as pd

df1= pd.read_csv("train-1.csv")
df2= pd.read_csv("train-2.csv")
df3= pd.read_csv("train-3.csv")
df4= pd.read_csv("train-4.csv")
df5= pd.read_csv("train-5.csv")
df6= pd.read_csv("train-6.csv")
df7= pd.read_csv("train-7.csv")
df8= pd.read_csv("train-8.csv")

df_final = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])

df_final.to_csv("train.csv", index=False)

train = pd.read_csv("train.csv")
df_train = train.drop(['Unnamed: 0'], axis=1)

df_train

#ids = df_train["tconst"]
#df_train[ids.isin(ids[ids.duplicated()])].sort_values("tconst")
```

```python
import json
from operator import itemgetter
import csv

f = open('writing.json', "r")
data = json.loads(f.read())

print(data[0]['movie'])
print(data[0])

clean_list = []

for line in data:
    clean_list.append(list(itemgetter('movie', 'writer')(line)))
clean_list


with open("write_clean.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(clean_list)

df = pd.read_csv('write_clean.csv')
df.columns = ['tconst', 'writer']
df.to_csv('write_clean2.csv')
```

```python
writers = pd.read_csv('write_clean2.csv')
writers
```

```python
df3 = df_train.merge(writers, how='left')
df3
```

```python
df_writer_test = df3.groupby(['tconst','primaryTitle', 'originalTitle', 'startYear', 'endYear', 'runtimeMinutes', 'numVotes', 'label'])['writer'].apply(', '.join).reset_index()
df5 = df3['writer'].groupby([df3.tconst,df3.primaryTitle, df3.originalTitle, df3.startYear, df3.endYear, df3.runtimeMinutes, df3.numVotes, df3.label]).apply(list).reset_index()
df_writer_test
```

```python
z = open('directing.json', "r")
director = json.loads(z.read())
dfmovie = director['movie']
dfmovie

dfdir = director['director']
dfdir
```

```python
mov_df = pd.DataFrame(dfmovie.items())
mov_df.columns = ['id', 'tconst']
dir_df = pd.DataFrame(dfdir.items())
dir_df.columns = ['id', 'director']
director_df = dir_df.merge(mov_df, on='id')
director_df.drop('id', axis=1, inplace=True)

director_df = director_df.groupby(['tconst'])['director'].apply(', '.join).reset_index()

director_df
```

```python
df_total = df_writer_test.merge(director_df)

df_total
```

```python
df_total.to_csv('training_full.csv')
```
